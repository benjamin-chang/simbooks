```python
def game(wager, odds=0.5, payout=1.0):
    if rand() < odds:
        return payout * wager
    else:
        return -wager
```

```python
def play(wager, odds=0.5, payout=1.0, bankroll=100, count=10, game=game, wager_rate=None):
    for j in range(count):
        if bankroll <= 0:
            return 0
        if wager_rate is not None:
            wager = wager_rate * bankroll
        bankroll += game(wager=min(bankroll, wager), payout=payout, odds=odds)
    return bankroll
```

```python
def sim(wager, odds=0.5, payout=1.0, iterations=10000, bankroll=100, count=10, game=game, wager_rate=None):
    return array([play(wager, payout=payout, odds=odds, bankroll=bankroll, count=count, game=game, wager_rate=wager_rate) for j in range(iterations)])
```

```python
wagers = linspace(0,1,101)
medians, means, stds = zip(*[(median(sims), mean(sims), std(sims)) for sims in (sim(wager=None, wager_rate=wager, odds=.50, payout=5, iterations=2000, count=200) for wager in wagers)])
means=array(means)
stds=array(stds)
plot(wagers, medians)
xlabel('wager')
ylabel('median outcome')
```

```python
plot(wagers, log1p(medians))
```

```python
plot(wagers, means+stds)
```

```python
plot(wagers, means-stds)
```
